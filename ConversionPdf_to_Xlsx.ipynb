```
# Se importa la libreria
from tabula import read_pdf
# importamos pandas
import pandas as pd
```

```
# Montamos el pdf que vamos a convertir
pdf = read_pdf("/content/drive/MyDrive/Convertir PDF a Excel con pandas/pdf/Libro1.pdf")
```

```
# Convertimos en DataFrame
df = pd.DataFrame(pdf[0])
# Asignamos nombre a las columnas
headers = ['Mes','Dia','Año','PptSalpo','TempMax','TempMin','Ppt','Wind','Hum','Solar']
df.columns = headers
```

```
# Validamos el DataFrame
df
```

```
# Eliminamos al fila 0
df = df.drop([0],axis=0)
```

```
# Validamos el cambio
df
```

```
# Exportamos a Excel
df.to_excel('/content/drive/MyDrive/Convertir PDF a Excel con pandas/pdf/Converion-PDF.xlsx')
```
